### Handling missing data

```python
import numpy as np
from sklearn.impute import SimpleImputer

numeric_cols = df.select_dtypes(include=[np.number]).columns
numeric_cols = df[numeric_cols].columns[~df[numeric_cols].isna().all()]

imputer = SimpleImputer(strategy='mean')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])

```

```python
# 1. Drop junk columns (if not already done)
df = df.drop(columns=["Unnamed: 0", "match report", "notes"], errors="ignore")

# 2. Define features and target
X_raw = df.drop(columns=["result"])
y = df["result"]
```

```python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(exclude=[np.number]).columns
```

```python
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean"))
])

categorical_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
ct = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_cols),
        ("cat", categorical_pipeline, categorical_cols)
    ]
)
```

```python
X = ct.fit_transform(df)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

model = Pipeline(steps=[
    ("preprocess", ct),
    ("classifier", LogisticRegression(max_iter=1000, multi_class="auto"))
])
```
